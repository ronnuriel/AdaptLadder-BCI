# T15 Session Drift Map

Computes train-split session drift metrics and figures.

```python
%load_ext autoreload
%autoreload 2
from pathlib import Path
import sys
ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT))
from src.drift_metrics import compute_session_stats, drift_metric_table, pairwise_mean_shift, session_mean_pca
from src.plotting import plot_drift_over_time, plot_mean_shift_heatmap, plot_pca_sessions

tables = ROOT / 'results/tables'
figures = ROOT / 'results/figures'
tables.mkdir(parents=True, exist_ok=True)
figures.mkdir(parents=True, exist_ok=True)

stats = compute_session_stats(ROOT / 'data/raw/hdf5_data_final', splits=('train',))
metrics = drift_metric_table(stats)
metrics.to_csv(tables / 't15_session_drift_metrics.csv', index=False)
pca = session_mean_pca(stats)
pca.to_csv(tables / 't15_session_mean_pca.csv', index=False)
sessions, matrix = pairwise_mean_shift(stats)
plot_drift_over_time(metrics, figures / 't15_drift_over_time.png')
plot_pca_sessions(pca, figures / 't15_pca_sessions.png')
plot_mean_shift_heatmap(sessions, matrix, figures / 't15_mean_shift_heatmap.png')
metrics.head()
```
